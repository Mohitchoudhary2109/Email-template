```python
box_content = []
for i in range(0,len(box)):
    box_content.append(box[i].get_text())
    
print('Number of Box Contents : ', len(box_content))  
box_content
```

```python
body_content = []
for i in range(0,len(body)):
    body_content.append(body[i].get_text())
    
print('Number of Body Contents : ', len(body_content))  
body_content
```

```python
box_content[:] = [box.lstrip('\n') for box in box_content]
box_content[:] = [box.rstrip('\n') for box in box_content]
print('Number of Box Contents : ', len(box_content))  
box_content
```

```python
body_content[:] = [body.lstrip('\n') for body in body_content]
body_content[:] = [body.rstrip('\n') for body in body_content]
print('Number of Body Contents : ', len(body_content))  
body_content
```

```python
df = pd.DataFrame()
df['Email subject']=box_content
df['content']=body_content

df
```

```python
df.to_csv('emailnewslettertemplates.csv',index=True)
```
